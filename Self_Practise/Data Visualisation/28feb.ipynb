```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2

ax = plt.axes(projection = '3d')
ax.plot_surface(xx, yy, z)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2

ax = plt.axes(projection = '3d')
ax.plot_surface(xx, yy, z, cmap = 'viridis')
```

```python
a = np.random.uniform(20, 30, 40)
b = np.random.uniform(20, 30, 40)
c = np.random.uniform(20, 30, 40)
d = np.random.uniform(20, 30, 40)
```

```python
years = [2010+i for i in range(10)]
years
```

```python
plt.stackplot(years, a, b, c, d, labels =['a','b','c','d'] )
plt.legend()
plt.xticks(years)
```

```python
#lineplot  =>line + std deviation 
```

```python
sns.lineplot(data = iris, x = 'sepal_length', y = 'petal_length', color = 'red') #seaborn expects the data in dataframe
```

```python
sns.lineplot(data = iris,x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```python
sns.set() #background
sns.lineplot(data = iris,x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```python
#reg plot, lmplot

sns.regplot(data = iris,  x = 'sepal_length',y = 'petal_length',  ci = None) #ci = to remove line shade
```

```python
sns.lmplot(data = iris,  x = 'sepal_length',y = 'petal_length',  ci = None, hue = 'species')
```

```python
sns.lmplot(data = iris,  x = 'sepal_length',y = 'petal_length',  ci = None, hue = 'species', col = 'species')
```

```python
#box plot
sns.boxplot(data = iris)
```

```python
#count plot
sns.countplot(data = iris)
```

```python
corr = iris.corr() #correletion between values range = -1 to 1 
corr
#0 -> no correletion
#1 -> strong +ve corr
#-1 -> strong -ve corr
```

```python
#pairplot
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
sns.pairplot(iris, hue = 'species', vars = ['sepal_width','sepal_length'])
```
